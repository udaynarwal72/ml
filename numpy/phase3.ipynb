```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== Zomato sales analysis ===")
print("\n sales data shape",sales_data.shape)
print("\n Sample data for 1st 3 restaurant: ",sales_data[0:3])
# to write it in the fancy language we will remove zero from the above line
print("\n Sample data for 1st 3 restaurant: ",sales_data[:3])
print("\n Sample data for 1st 3 restaurant: ",sales_data[:,1:])
```

```python
# total sales per year
print(np.sum(sales_data,axis=0))
# zero for the column
# one for the row
yearly_total = np.sum(sales_data[:,1:],axis = 0)
print("Yearly total",yearly_total)

# minimum sales per retaurant
min_sales = np.min(sales_data[:,1:],axis = 1)
print("Minimum sales",min_sales)

# maximum sales per year
max_sale = np.max(sales_data[:,1:],axis = 0)
print("max sales per year",max_sale)

# average sales per restaurant
average_sales = np.mean(sales_data[:,1:],axis=1)
print("avg sales",average_sales)
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis = 0))

plt.title("Average cumulative sales accross all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show() # This will be used in jnb only

```

### vector multiplication

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition: ",vector1+ vector2)

print("\n vector mulitplication",vector1*vector2)

print("\n Dot product",np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("Angle",angle)
```

```python
restaurant_types = np.array(['biryani','chinese','pizza','burger','cafe'])

# This means every elements of the restaurant will get iterated
vectorized_upper = np.vectorize(str.upper)

print("Vectorized upper",vectorized_upper)
print("Vectorized upper",vectorized_upper(restaurant_types))
```
